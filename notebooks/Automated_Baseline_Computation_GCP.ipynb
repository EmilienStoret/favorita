```python
from favorita import functions
```

```python
d={'store':[], 'product': [], 'baseline_score': []}
for s in store:
    for p in product:
        path = f'/home/emilien/code/EmilienStoret/favorita/Data/data_pulled_from_GCP/store'+str(s)+'_'+str(p)
        df = pd.read_csv(path)
        baseline_score = functions.get_baseline_score(df)
        d['store'].append(s)
        d['product'].append(p)
        d['baseline_score'].append(baseline_score)
df_baseline = pd.DataFrame(d)
df_baseline
```
